#### Ejercicio 19. ¿Qué puedes decir sobre el precio por metro cuadrado (precio/m2) entre los municipios de 'Getafe' y 'Alcorcón'? Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)

Pista: Crea una nueva columna llamada `pps` (price per square en inglés) y luego analiza los valores

```python
df_cinturon_sur = df[df['level5'].isin(['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón'])].copy()

df_cinturon_sur.loc[:, 'pps'] = df_cinturon_sur['price'] / df_cinturon_sur['surface']

getafe = df_cinturon_sur[df_cinturon_sur['level5'] == 'Getafe']
alcorcon = df_cinturon_sur[df_cinturon_sur['level5'] == 'Alcorcón']

getafe_pps_mean = getafe['pps'].mean()
alcorcon_pps_mean = alcorcon['pps'].mean()

print(f"Precio por metro cuadrado promedio en Getafe: {getafe_pps_mean}")
print(f"Precio por metro cuadrado promedio en Alcorcón: {alcorcon_pps_mean}")

#TODO: El precio por metro cuadrado en Alcorcón es más alto que en Getafe. Esto puede ser un indicio de que las viviendas en Alcorcón tienden a ser más caras por metro cuadrado, posiblemente debido a factores como la ubicación, la demanda, o características de las viviendas en cada municipio.
```

#### Ejercicio 21. Realiza un trazado de las coordenadas (columnas latitud y longitud) del cinturón sur de Madrid por color de cada población (debes usar el subconjunto obtenido del Ejercicio 14) (★★★★)

Ejecuta la siguiente celda y luego comienza a codear en la siguiente. Debes implementar un código simple que transforme las columnas de coordenadas en un diccionario de Python (agrega más información si es necesario) y agrégala al mapa.

```python
from ipyleaflet import Map, basemaps

# Crear un mapa centrado en Madrid (latitud 40.35, longitud -3.75)
map = Map(center=(40.35, -3.75), zoom=10, min_zoom=1, max_zoom=20, basemap=basemaps.OpenStreetMap.Mapnik)
# Mostrar el mapa
map
```

```python
## Aquí: traza la coordenadas de los estados

## PON TU CÓDIGO AQUÍ:
import pandas as pd
from ipyleaflet import Map, Marker, MarkerCluster, Icon

# Leer el archivo CSV
df = pd.read_csv('assets/real_estate.csv', sep=';')

# Filtrar los datos para el cinturón sur
cinturon_sur = df[df['level5'].isin(['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón'])].copy()

# Asegurarse de que las columnas 'latitude' y 'longitude' sean numéricas
cinturon_sur['latitude'] = pd.to_numeric(cinturon_sur['latitude'], errors='coerce')
cinturon_sur['longitude'] = pd.to_numeric(cinturon_sur['longitude'], errors='coerce')

# Eliminar filas con valores nulos en latitud o longitud
cinturon_sur = cinturon_sur.dropna(subset=['latitude', 'longitude'])

# Definir colores para cada población
colors = {
    'Fuenlabrada': 'red',
    'Leganés': 'blue',
    'Getafe': 'green',
    'Alcorcón': 'purple'
}

# Crear los marcadores
markers = []
for _, row in cinturon_sur.iterrows():
    icon = Icon(color=colors.get(row['level5'], 'black'), icon_color='white', icon='info-sign', draggable=False)
    marker = Marker(location=(row['latitude'], row['longitude']), title=row['level5'], icon=icon)
    markers.append(marker)

# Crear el mapa centrado en Madrid
map = Map(center=(40.35, -3.75), zoom=10)

# Crear un clúster de los marcadores
marker_cluster = MarkerCluster(markers=markers)
map.add_layer(marker_cluster)

# Mostrar el mapa
map
```
